```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout, Rescaling

# Image preprocessing
img_size = (224, 224)
batch_size = 32

# Load datasets
training_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "sample_data/datasets/training",
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)

validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "sample_data/datasets/validation",
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)

testing_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "sample_data/datasets/testing",
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)

# Extract class names before mapping
class_names = training_dataset.class_names
num_classes = len(class_names)
print("Detected classes:", class_names)

# Normalize images
normalization_layer = Rescaling(1./255)
training_dataset = training_dataset.map(lambda x, y: (normalization_layer(x), y))
validation_dataset = validation_dataset.map(lambda x, y: (normalization_layer(x), y))
testing_dataset = testing_dataset.map(lambda x, y: (normalization_layer(x), y))

# Build MLP model
model = Sequential()
model.add(Flatten(input_shape=(224, 224, 3)))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

# Compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train model
model.fit(training_dataset, validation_data=validation_dataset, epochs=10, batch_size=batch_size)

# Evaluate model
test_loss, test_acc = model.evaluate(testing_dataset)
print(f'Test accuracy: {test_acc:.2f}')
```

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Flatten, Dense, Dropout, Rescaling, BatchNormalization

# Image settings
img_size = (224, 224)
batch_size = 32

# Load datasets
training_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "sample_data/datasets/training",
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)
validation_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "sample_data/datasets/validation",
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)
testing_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    "sample_data/datasets/testing",
    image_size=img_size,
    batch_size=batch_size,
    label_mode='categorical'
)

# Save class names before mapping
class_names = training_dataset.class_names
num_classes = len(class_names)

# Normalize images
normalization_layer = Rescaling(1./255)
training_dataset = training_dataset.map(lambda x, y: (normalization_layer(x), y))
validation_dataset = validation_dataset.map(lambda x, y: (normalization_layer(x), y))
testing_dataset = testing_dataset.map(lambda x, y: (normalization_layer(x), y))

# Build improved MLP
model = Sequential()
model.add(Flatten(input_shape=(224, 224, 3)))
model.add(Dense(1024, activation='relu'))
model.add(BatchNormalization())
# model.add(Dropout(0.3))
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))

# Compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train full 20 epochs without early stopping
model.fit(training_dataset, validation_data=validation_dataset, epochs=10, batch_size=batch_size)

# Evaluate
test_loss, test_acc = model.evaluate(testing_dataset)
print(f'Test accuracy: {test_acc:.2f}')
```
